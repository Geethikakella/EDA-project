# REMOVING OUTLIERS USING IQR METHOD

```python
len(df)
```

```python
q1 = df["YearsAtCompany"].quantile(0.25)
q3 = df["YearsAtCompany"].quantile(0.75)
iqr = q3-q1
```

```python
q1,q3,iqr
```

```python
upper_limit = q3 + (1.5 * iqr)
lower_limit = q3 - (1.5 * iqr)
upper_limit,lower_limit
```

```python
#trimming - delete the outlier data
df1 = df.loc[(df["YearsAtCompany"] < upper_limit) & (df["YearsAtCompany"] > lower_limit)]
print(len(df))
print(len(df1))
print(len(df)-len(df1))
```

```python
sns.boxplot(df1["YearsAtCompany"],orient="h")
```

```python
#capping -change the outlier value to upper (or) lower limit value
df1 = df.copy()
df1.loc[(df["YearsAtCompany"]>upper_limit), "YearsAtCompany"] = upper_limit
df1.loc[(df["YearsAtCompany"]<lower_limit), "YearsAtCompany"] = lower_limit
```

```python
sns.boxplot(df1["YearsAtCompany"],orient="h")
```
